## Problem Statement:

A bike-sharing system is a service in which bikes are made available for shared use to individuals on a short term basis for a price or free. Many bike share systems allow people to borrow a bike from a "dock" which is usually computer-controlled wherein the user enters the payment information, and the system unlocks it. This bike can then be returned to another dock belonging to the same system.


A US bike-sharing provider BoomBikes has recently suffered considerable dips in their revenues due to the ongoing Corona pandemic. The company is finding it very difficult to sustain in the current market scenario. So, it has decided to come up with a mindful business plan to be able to accelerate its revenue as soon as the ongoing lockdown comes to an end, and the economy restores to a healthy state. 


In such an attempt, BoomBikes aspires to understand the demand for shared bikes among the people after this ongoing quarantine situation ends across the nation due to Covid-19. They have planned this to prepare themselves to cater to the people's needs once the situation gets better all around and stand out from other service providers and make huge profits.


They have contracted a consulting company to understand the factors on which the demand for these shared bikes depends. Specifically, they want to understand the factors affecting the demand for these shared bikes in the American market. 

The company wants to know:

    - Which variables are significant in predicting the demand for shared bikes.
    - How well those variables describe the bike demands
    
Based on various meteorological surveys and people's styles, the service provider firm has gathered a large dataset on daily bike demands across the American market based on some factors. 

**So interpretation is important!**

## Understanding & Preparation  & Visualize the Data

### dteday : date

```python
df_day.dteday
```

```python
# Converting date object to datetime format
df_day.dteday = df_day.dteday.astype('datetime64')
df_day.dteday
```

```python
years = df_day.dteday.dt.year
len(np.unique(years))
```

```python
# func for plot bar graph and show values over the graph
def plotBarGraph(series):
    vc = series.value_counts()
    vc.plot.barh()

    for i, v in enumerate(vc):
        plt.text(v, i , str(v), color='green')

    plt.show()   
```

### season : season (1:spring, 2:summer, 3:fall, 4:winter)

```python
df_day.season.value_counts()
```

```python
def transform_season(x):
    return x.map({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
```

```python
df_day[['season']] = df_day[['season']].apply(transform_season)
```

```python
df_day.season.value_counts()
```

```python
plotBarGraph(df_day.season)
```

### yr : year (0: 2018, 1:2019)

```python
plotBarGraph(df_day.yr)
```

### mnth : month ( 1 to 12)

```python
def transform_month(x):
    return x.map({1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})
```

```python
df_day[['mnth']] = df_day[['mnth']].apply(transform_month)
```

```python
plotBarGraph(df_day.mnth)
```

### holiday : weather day is a holiday or not (extracted from http://dchr.dc.gov/page/holiday-schedule)
#### 0 - No Holiday
#### 1 - Holiday

```python
plotBarGraph(df_day.holiday)
```

### weekday : day of the week

```python
def transform_week(x):
    return x.map({1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat', 7: 'Sun'})
```

```python
df_day[['weekday']] = df_day[['weekday']].apply(transform_week)
```

```python
plotBarGraph(df_day.weekday)
```

### workingday : if day is neither weekend nor holiday is 1, otherwise is 0.

```python
plotBarGraph(df_day.workingday)
```

### weathersit :
    1: Clear, Few clouds, Partly cloudy, Partly cloudy
    2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
    3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
    4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

```python
def transform_weather(x):
    return x.map({1: 'Clear', 2: 'Mist', 3: 'Rain', 4: 'Snow'})
```

```python
df_day[['weathersit']] = df_day[['weathersit']].apply(transform_weather)
```

```python
plotBarGraph(df_day.weathersit)

# Above we have 4 categories but here in graph showing 3, because 4 is the part of 1, 2 & 3.
```

## Step 3: Data Preparation

```python
df_day.head()
```

```python
df_day_cat = df_day[['season', 'mnth', 'weekday', 'weathersit']]
df_day_cat.head()
```

```python
df_day_dummies = pd.get_dummies(df_day_cat, drop_first=True)
df_day_dummies.head()
```

```python
df_day_cat.columns
```

```python
df_day = df_day.drop(['instant', 'season', 'mnth', 'weekday', 'weathersit','dteday'], axis=1)
df_day.head()
```

```python
df_day = pd.concat([df_day, df_day_dummies], axis=1)
df_day.head()
```

## Step 4: Splitting the Data into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

np.random.seed(0)
df_train, df_test = train_test_split(df_day, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
df_train.info()
```

```python
df_test.info()
```

#### Rescaling the Features 

```python
from sklearn.preprocessing import MinMaxScaler 

scaler = MinMaxScaler()

features = ['temp', 'hum', 'windspeed','casual','registered','cnt']

df_train[features] = scaler.fit_transform(df_train[features])
```

```python
df_train.describe()
```

```python
plt.figure(figsize = (35, 25))
sns.heatmap(df_train.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

```python
plt.scatter(df_train.temp, df_train.cnt)
plt.show()
```

#### Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('cnt')
X_train = df_train.drop(["casual","registered"],axis=1) 
```

```python
X_train.shape
```

## Step 5: Building a linear model

```python
import statsmodels.api as sm

# Add a constant
X_train_lm = sm.add_constant(X_train)

# Create fit model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# Check the parameters obtained
lr.params
```

```python
lr.summary()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
rfe = RFE(lm, 18)

rfe.fit(X_train, y_train)
```

```python
rfe.ranking_
```

```python
rfe.support_
```

```python
cols = X_train.columns[rfe.support_]

X_train_rfe = X_train[cols]

# Add a constant
X_train_rfe = sm.add_constant(X_train_rfe)
X_train_rfe.head()
```

```python
X_train_rfe = X_train_rfe.drop(['const'],axis=1)
X_train_rfe.shape
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit()   
lm.summary()
```

```python
lm = LinearRegression()
rfe = RFE(lm, 10)

rfe.fit(X_train, y_train)
```

```python
cols = X_train.columns[rfe.support_]

X_train_rfe = X_train[cols]

# Add a constant
X_train_rfe = sm.add_constant(X_train_rfe)
X_train_rfe.head()
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit()   
lm.summary()
```

```python
lm = LinearRegression()
rfe = RFE(lm, 6)

rfe.fit(X_train, y_train)
```

```python
cols = X_train.columns[rfe.support_]

X_train_rfe = X_train[cols]

# Add a constant
X_train_rfe = sm.add_constant(X_train_rfe)
X_train_rfe.head()
```

```python
lm = sm.OLS(y_train, X_train_rfe).fit()   
lm.summary()
```

#### Checking VIF

```python
# Check for the VIF values of the feature variables. 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
X_train_rfe = X_train_rfe.drop(['const'],axis=1)
X_train_rfe
```

```python
features = ['temp', 'season_Spring','hum','windspeed','yr','weathersit_Rain']

# Heatmap
sns.heatmap(df_day[features].corr(), cmap="YlGnBu", annot=True)
plt.show()
```

## Step 7: Residual Analysis of the train data

```python
y_train_cnt = lr.predict(X_train_lm)
y_train_cnt.shape
```

```python
# Plot the histogram of the error terms
fig = plt.figure()
sns.distplot((y_train - y_train_cnt), bins = 20)
fig.title('Error Terms', fontsize = 20)                  # Plot heading 
plt.xlabel('Errors', fontsize = 18)                         # X-label
```

## Step 8: Making Predictions Using the Final Model

#### Applying the scaling on the test sets 

```python
df_test[features] = scaler.fit_transform(df_test[features])
df_test
```

```python
y_test = df_test.pop('cnt')
X_test = df_test.drop(["casual","registered"],axis=1)
X_test.head()
```

```python
X_test_rfe = X_test[features]
X_test_rfe
```

```python
# Add a constant to the test set created
X_test_rfe = sm.add_constant(X_test_rfe)
```

```python
X_test_rfe
```

```python
# Making predictions
y_pred = lm.predict(X_test_rfe)
y_pred.shape
```

## Step 9: Model Evaluation

Let's now plot the graph for actual versus predicted values.

```python
# Plotting y_test and y_pred to understand the spread

fig = plt.figure()
plt.scatter(y_test, y_pred)
fig.suptitle('y_test vs y_pred', fontsize = 20)              # Plot heading 
plt.xlabel('y_test', fontsize = 18)                          # X-label
plt.ylabel('y_pred', fontsize = 16)      
```

```python
# r2_score for 6 variables
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
